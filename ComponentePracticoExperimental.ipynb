# Cargar el conjunto de datos
## Muestre los datos cargados

```python
import matplotlib.pyplot as plt  
import seaborn as sns
import pandas as pd
import numpy as np 
from sklearn import preprocessing
datos = pd.read_csv('compras.csv',delimiter=',')
datos.head(20)
```

# Muestre los tipos de datos

```python
tipos = datos.columns.to_series().groupby(datos.dtypes).groups
```

```python
# Armando lista de columnas categóricas
ctext = tipos[np.dtype('object')]
len(ctext)  # cantidad de columnas con datos categóricos. 
```

```python
# Armando lista de columnas numéricas
columnas = datos.columns  # lista de todas las columnas
cnum = list(set(columnas) - set(ctext))
len(cnum)
```

# Muestre los estadisticos basico

```python
datos.describe()
```

# Utilice la media (al entero mas cercano) como valor para imputar los datos faltantes

```python
datos.count()
```

```python
for c in cnum:
    mean = datos[c].mean()
    datos[c] = datos[c].fillna(mean)

datos.isnull().any().any()
datos.count()
```
